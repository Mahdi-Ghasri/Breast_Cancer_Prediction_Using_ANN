  <h2>🔴دسته‌بندی ویژگی‌ها در دیتاست</h2>
  <font size="4">
      ویژگی‌های این دیتاست مربوط به بررسی بافت و ساختار هسته‌ی سلول‌های سینه در تصاویر میکروسکوپی هستند. این ویژگی‌ها از نظر آماری و هندسی در سه گروه زیر طبقه‌بندی شده‌اند که هر گروه می‌تواند در تشخیص خوش‌خیم یا بدخیم بودن توده‌ها نقش کلیدی داشته باشد.<br>
      <li>
        <b>ویژگی‌های میانگین (Mean Features):</b>
        <br>
        این گروه از ویژگی‌ها میانگین اندازه‌گیری‌های مربوط به هر نمونه را توصیف می‌کنند. مثلاً <code>radius_mean</code> میانگین شعاع هسته‌هاست و <code>texture_mean</code> میانگین بافت تصویر است. این مقادیر نمای کلی از ساختار فیزیکی توده را ارائه می‌دهند.
        در بحث‌های هندسی، این ویژگی‌ها معمولاً با مشخصه‌های اصلی شکل هندسی (مساحت، محیط، شعاع، یکنواختی، پیچیدگی مرز و ...) مرتبط هستند.
      </li>
      <br>
      <li>
        <b>ویژگی‌های خطای استاندارد (Standard Error Features):</b>
        <br>
        این دسته نشان‌دهنده‌ی میزان نوسان یا پراکندگی اندازه‌گیری‌ها برای هر ویژگی هستند. مثلاً <code>radius_se</code> مقدار انحراف استاندارد اندازه شعاع‌ها در نقاط مختلف سلول است.
        این ویژگی‌ها به شما اطلاعاتی در مورد <u>پایداری</u> ساختار یا تغییرپذیری شکل سلول می‌دهند. در پردازش هندسی، نوسان بالا ممکن است نشانگر لبه‌های نامنظم یا تغییرات غیرنرمال در بافت باشد.
      </li>
      <br>
      <li>
        <b>ویژگی‌های بزرگ‌ترین مقدار (Worst / Largest Features):</b>
        <br>
        این گروه نماینده‌ی بیشترین مقدار مشاهده‌شده برای هر ویژگی است (مثلاً <code>radius_worst</code> بزرگ‌ترین شعاع مشاهده‌شده در مرز سلول).
        این مقادیر معمولاً نشان‌دهنده‌ی حداکثر انحراف از حالت عادی هستند و در طبقه‌بندی توده‌های بدخیم که تمایل به رشد نامنظم دارند، بسیار مهم‌اند.
        در هندسه‌ی محاسباتی، این ویژگی‌ها می‌توانند نمایانگر نقاطی با بیش‌ترین تحدب، پیچیدگی یا بی‌نظمی باشند.
      </li>
      <br><br>
      <font size=4 color='red'>جمع‌بندی</font>: در فرآیند Feature Engineering، درک دقیق این سه دسته ویژگی به شما کمک می‌کند تا بتوانید با ترکیب آن‌ها، ویژگی‌های جدید هندسی بسازید (مانند نسبت حداکثر به میانگین، ضریب نوسان، نسبت سطح به محیط، و غیره) که به مدل در تشخیص دقیق‌تر کمک خواهند کرد.

        <li>هر بخش از تابع را با دقت تکمیل کنید تا خروجی‌های واضح و مفیدی تولید کند.</li>

        <li>در پایان اجرا، خروجی باید نمایی خلاصه و قابل فهم از وضعیت فعلی دیتاست ارائه دهد.</li>
    <h3><strong>نکته پایانی:</strong></h3>
        <font size=4>
        این نوع توابع برای هر پروژه‌ی داده‌کاوی یا یادگیری ماشین به‌عنوان یک گام ابتدایی ضروری هستند. با تسلط بر این بررسی‌ها، شما درک عمیق‌تری از داده‌های خام پیدا خواهید کرد و تصمیمات بهتری در مراحل بعدی می‌گیرید.

```python
# TODO : تابع زیر را کامل کنید
def explore_dataset(df):
    """
    انجام بررسی اولیه روی یک دیتافریم pandas.

    پارامترها:
    ----------
    df : pd.DataFrame
        دیتافریم ورودی برای بررسی.

    عملکرد:
    -------
    - نمایش خلاصه‌ای از اطلاعات ساختاری دیتافریم
    - بررسی مقادیر گمشده
    - بررسی ردیف‌های تکراری
    - نمایش تعداد مقادیر یکتا در هر ستون
    - نمایش نوع داده‌ای هر ستون
    """



    # 1. اطلاعات کلی دیتافریم
    print("Dataframe info\n")
    print(df.info())
    print("-" * 50)

    # 2. بررسی مقادیر گمشده
    print("\n[2] Missing values per column:")
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    if not missing.empty:
        print(missing)
    else:
        print("✅ هیچ داده‌ی گمشده‌ای وجود ندارد.")
    print("-" * 50)

    # 3. بررسی ردیف‌های تکراری
    num_duplicates = df.duplicated().sum()
    print(f"\n[3] Number of duplicated rows: {num_duplicates}")
    print("-" * 50)

    # 4. تعداد مقادیر یکتا در هر ستون
    print("\n Number of unique values per column:")
    print(df.nunique().sort_values(ascending=False))
    print("-" * 50)

    # 5. نوع داده‌ای هر ستون
    print("\n Data types of columns:")
    print(df.dtypes)
```

```python
explore_dataset(df)
```

```python
# داده های عددی
numerical_cols = df.drop(columns=['diagnosis']).columns.tolist()
```

```python
# TODO: متد زیر را برای نشان دادن هیستوگرام تکمیل و متغیر های ذکر شده را بررسی کنید.
def plot_numerical_histograms(df, numerical_cols, bins=30):
    """
    رسم هیستوگرام برای ستون‌های عددی با استفاده از subplot.

    پارامترها:
    - df: دیتافریم ورودی
    - numerical_cols: لیست ویژگی‌های عددی
    - bins: تعداد سطل‌های هیستوگرام
    """
    n = len(numerical_cols)
    cols = 4  # تعداد ستون‌ها در subplot
    rows = int(np.ceil(n / cols))  # محاسبه تعداد ردیف‌ها

    fig, axes = plt.subplots(rows, cols, figsize=(20, 4 * rows))
    axes = axes.flatten()

    for i, col in enumerate(numerical_cols):
        axes[i].hist(df[col], bins=bins, color='skyblue', edgecolor='black')
        axes[i].set_title(col, fontsize=10)
        axes[i].grid(True)

    # اگر subplot اضافه داریم، خاموششان کن
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')

    plt.tight_layout()
    plt.show()
```

```python
# TODO: از تابع نوشته استفاده کنید و هیستوگرام ویژگی های عددی را رسم کنید.
plot_numerical_histograms(df, numerical_cols, bins=30)
```

            برای ویژگی‌هایی مانند radius, area, perimeter که اندازه‌های هندسی داده‌ها هستند، نسبت‌های مختلف بین این اندازه‌ها می‌توانند اطلاعات مهمی درباره شکل داده‌ها بدهند. برای مثال، نسبت area_mean به perimeter_mean می‌تواند ویژگی‌های جدید و مفیدی بسازد.

```python
# TODO: نسبت‌های هندسی بین اندازه‌ها  را بدست آورید
# مثال :
df_copy['area_to_perimeter'] = df_copy['area_mean'] / df_copy['perimeter_mean']

# نسبت مساحت به شعاع مربع - در حالت ایده‌آل برای دایره باید تقریباً ثابت باشد
df_copy['area_to_radius2'] = df_copy['area_mean'] / (df_copy['radius_mean'] ** 2)

# نسبت محیط به شعاع - بررسی تطابق با شکل دایره‌ای (فرمول کلاسیک: 2πr ≈ perimeter)
df_copy['perimeter_to_radius'] = df_copy['perimeter_mean'] / df_copy['radius_mean']

# نسبت محیط به مساحت - رابطه معکوس با تراکم
df_copy['perimeter_to_area'] = df_copy['perimeter_mean'] / df_copy['area_mean']
```

<div dir="rtl">
    <h2>تفاضل بین نسخه‌های worst و mean (Worst - Mean Differences)<h2>
    <font size=4>
        <li>
            برای ویژگی‌هایی مانند radius, area, perimeter که اندازه‌های هندسی داده‌ها هستند، نسبت‌های مختلف بین این اندازه‌ها می‌توانند اطلاعات مهمی درباره شکل داده‌ها بدهند. برای مثال، نسبت area_mean به perimeter_mean می‌تواند ویژگی‌های جدید و مفیدی بسازد.
        </li>
    </font>
</div>

```python
# TODO : محاسبه اختلاف بین متغیر های mean و worst
# مثال:
df_copy['smoothness_diff'] = df_copy['smoothness_worst'] - df_copy['smoothness_mean']

# TODO: برای بقیه ویژگی ها همانند کد بالا بنویسید. در انتخاب ویژگی طبق توضیحات عمل کنید.
# شعاع
df_copy['radius_diff'] = df_copy['radius_worst'] - df_copy['radius_mean']

# بافت
df_copy['texture_diff'] = df_copy['texture_worst'] - df_copy['texture_mean']

# محیط
df_copy['perimeter_diff'] = df_copy['perimeter_worst'] - df_copy['perimeter_mean']

# مساحت
df_copy['area_diff'] = df_copy['area_worst'] - df_copy['area_mean']

# فشردگی
df_copy['compactness_diff'] = df_copy['compactness_worst'] - df_copy['compactness_mean']

# فرورفتگی
df_copy['concavity_diff'] = df_copy['concavity_worst'] - df_copy['concavity_mean']

# نقاط فرورفته
df_copy['concave_points_diff'] = df_copy['concave points_worst'] - df_copy['concave points_mean']

# تقارن
df_copy['symmetry_diff'] = df_copy['symmetry_worst'] - df_copy['symmetry_mean']

# پیچیدگی مرز
df_copy['fractal_dimension_diff'] = df_copy['fractal_dimension_worst'] - df_copy['fractal_dimension_mean']
```

<div dir="rtl">
    <h2>ویژگی‌های شدت تغییرات نسبی (Relative Standard Error Features)<h2>
    <font size=4>
        <li>
           برای هر ویژگی، محاسبه نسبت انحراف معیار به میانگین (Standard Error / Mean) می‌تواند نشان دهد که تغییرات آن ویژگی در مقایسه با مقدار میانگین چقدر زیاد است. این ویژگی‌ها می‌توانند شامل radius_se / radius_mean یا area_se / area_mean باشن
        </li>
    </font>
</div>

```python
# TODO: محاسبه نسبت انحراف معیار به میانگی

# لیست ویژگی‌های اصلی بدون پسوند
features = [
    'radius', 'texture', 'perimeter', 'area', 'smoothness',
    'compactness', 'concavity', 'concave points',
    'symmetry', 'fractal_dimension'
]

# حلقه برای ساخت ویژگی‌های جدید بر اساس نسبت SE به Mean
for feature in features:
    rel_error_name = f'{feature}_rel_error'
    se_col = f'{feature}_se'
    mean_col = f'{feature}_mean'
    df_copy[rel_error_name] = df_copy[se_col] / df_copy[mean_col]
```

<div dir="rtl">
    <h2>میانگین هندسی ترکیبی ویژگی‌های با هم‌بستگی بالا (Composite Features)<h2>
    <font size=4>
        <li>
           اگر ویژگی‌هایی مانند compactness_mean و compactness_worst هم‌بستگی بالایی داشته باشند، می‌توانید آن‌ها را ترکیب کرده و یک ویژگی جدید بسازید، مثل میانگین هندسی یا میانگین ساده این ویژگی‌ها برای کاهش ابعاد و استفاده بهینه از اطلاعات.
        </li>

```python
# TODO: میانگین هندسی ترکیبی ویژگی‌های با هم‌بستگی بالا
# ایجاد میانگین هندسی ترکیبی برای ویژگی‌های highly correlated

df_copy['compactness_geom_mean'] = (df_copy['compactness_mean'] * df_copy['compactness_se'] * df_copy['compactness_worst']) ** (1/3)
df_copy['concavity_geom_mean'] = (df_copy['concavity_mean'] * df_copy['concavity_se'] * df_copy['concavity_worst']) ** (1/3)
df_copy['concave_points_geom_mean'] = (df_copy['concave points_mean'] * df_copy['concave points_se'] * df_copy['concave points_worst']) ** (1/3)
df_copy['radius_geom_mean'] = (df_copy['radius_mean'] * df_copy['radius_se'] * df_copy['radius_worst']) ** (1/3)
df_copy['perimeter_geom_mean'] = (df_copy['perimeter_mean'] * df_copy['perimeter_se'] * df_copy['perimeter_worst']) ** (1/3)
df_copy['area_geom_mean'] = (df_copy['area_mean'] * df_copy['area_se'] * df_copy['area_worst']) ** (1/3)
```

```python
# TODO: ترکیب میانگین و worst برای یک ویژگی نرمال‌شده
# مثال:
df_copy['avg_radius'] = (df_copy['radius_mean'] + df_copy['radius_worst']) / 2

# TODO: برای بقیه ویژگی ها همانند کد بالا بنویسید. در انتخاب ویژگی طبق توضیحات عمل کنید.

# ترکیب mean و worst برای چند ویژگی منتخب

df_copy['avg_smoothness'] = (df_copy['smoothness_mean'] + df_copy['smoothness_worst']) / 2
df_copy['avg_symmetry'] = (df_copy['symmetry_mean'] + df_copy['symmetry_worst']) / 2
df_copy['avg_concavity'] = (df_copy['concavity_mean'] + df_copy['concavity_worst']) / 2
df_copy['avg_compactness'] = (df_copy['compactness_mean'] + df_copy['compactness_worst']) / 2
df_copy['avg_concave_points'] = (df_copy['concave points_mean'] + df_copy['concave points_worst']) / 2
df_copy['avg_fractal_dimension'] = (df_copy['fractal_dimension_mean'] + df_copy['fractal_dimension_worst']) / 2
```

```python
# TODO: داده ها را مقیاس بندی کنید

# در این مسئله و برای مدل‌های ANN، بهترین انتخاب StandardScaler است، چون بسیاری از الگوریتم‌های گریدینت‌بیس مانند شبکه عصبی، وقتی داده‌ها حول ۰ پخش شده باشند بهتر آموزش می‌بینند.

# مقیاس‌گذاری
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# حذف NaN / inf بعد از Scaling (در صورت وجود)
X_train_scaled = np.nan_to_num(X_train_scaled)
X_test_scaled = np.nan_to_num(X_test_scaled)
```

```python
# TODO: طراحی شبکه عصبی
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers, initializers

def build_model(input_shape):
    model = models.Sequential([
        layers.Dense(64, activation='relu', input_shape=(input_shape,),
                     kernel_regularizer=regularizers.l2(0.0001),
                     kernel_initializer=initializers.HeNormal()),
        layers.Dropout(0.3),

        layers.Dense(32, activation='relu',
                     kernel_regularizer=regularizers.l2(0.0001),
                     kernel_initializer=initializers.HeNormal()),
        layers.Dropout(0.2),

        layers.Dense(1, activation='sigmoid')  # چون مسئله دوتاییه
    ])

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

model = build_model(X_train_scaled.shape[1])

```
